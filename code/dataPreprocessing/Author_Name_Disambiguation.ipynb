```python
import pandas as pd
import disamby.preprocessors as pre
from disamby import Disamby
```

```python
#Disambiguate authors names with names from the crawled data
data1 = pd.read_csv(r'/Users/nadee/Documents/1Graduate Studies/Term 2/NLP/Project/Code/authors_normalized_affiliations_replaced_some.csv') 
data2 = pd.read_csv(r'/Users/nadee/Documents/1Graduate Studies/Term 2/NLP/Project/Code/KFUPM_Faculty_Disambiguated.csv') 

dataFrame1 = pd.DataFrame(data1, columns=['paper','title', 'author', 'affiliation', 'cite', 'year'])
dataFrame2 = pd.DataFrame(data2, columns=['author', 'affiliation'])
```

```python
authors1 = dataFrame1["author"]
authors2 = dataFrame2["author"]

authors_list = authors1.tolist()
authors_list2 = authors2.tolist()

#Combine the two lists to find name disambiguation
authors_list.extend(authors_list2)

for x in range(len(authors_list)):
    authors_list[x] = authors_list[x].strip().title()
    
print(len(authors_list))
```

```python
df = pd.DataFrame({
    'author': authors_list},
)
```

```python
# define the pipeline to process the strings, note that the last step must return
# a tuple of strings
pipeline = [
    pre.normalize_whitespace,
    pre.remove_punctuation,
    lambda x: pre.trigram(x) + pre.split_words(x)
]
```

```python
dis = Disamby(df, pipeline)
```

```python
unique_authors = dis.disambiguated_sets(threshold=0.5)
```

```python
for x in unique_authors:
    if len(x)>1:
        #print('Length:', len(x), "Unique:", x)
        #print(x, type(x))
        s = list(x) 
        #print("s", s)
        for i in range(len(s)):
            if i == 0:
                authors_list[s[i]] = authors_list[s[i]]
            else:
                #print(s[i], authors_list[s[i]])
                print('Pre:', authors_list[s[i]])
                authors_list[s[i]] = authors_list[s[i-1]]
                print('Post:', authors_list[s[i]])
```

```python
authors1 = authors_list[:len(dataFrame1)]
authors2 = authors_list[len(dataFrame1):]
```

```python
from pandas import DataFrame

df = DataFrame(authors1,columns=['author'])
#print(df)
table1 = {"paper": [], "title": [], "author": [], "affiliation": [], "cite": [], "year": []}

for x in range(len(dataFrame1)):
    table1["paper"].append(dataFrame1['paper'][x])
    table1["title"].append(dataFrame1['title'][x])
    table1["author"].append(df['author'][x])
    table1["affiliation"].append(dataFrame1['affiliation'][x])
    table1["cite"].append(dataFrame1['cite'][x])
    table1["year"].append(dataFrame1['year'][x])
```

```python
df = DataFrame(authors2,columns=['author'])
#print(df)
table2 = {"author": [], "affiliation": []}

for x in range(len(dataFrame2)):
    table2["author"].append(df['author'][x])
    table2["affiliation"].append(dataFrame2['affiliation'][x])
```

```python
df=pd.DataFrame.from_dict(table1,orient='index').transpose()
df.to_csv('authors_disambiguation_normalized_affiliations_replaced_some.csv', index = False)
```
