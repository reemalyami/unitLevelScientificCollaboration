```python
#Counting authors with coll, School, rest inst, res ctr, ctr res excellence, res inst ri, cer res inst, OR centers res excellence affiliations
collSchoolCount = ctrCount = 0
for i in range(len(affiliations_list)):
    if(re.search(r'coll|\bGeneral Studies College\b|\bSchool\b|\bsch engn\b', affiliations_list[i])):
        collSchoolCount = collSchoolCount+1
        print('College/School:', affiliations_list[i])
    elif(re.search(r'\bres inst\b|\bres ctr\b|\bctr res excellence\b|\bres inst ri\b|\bcer res inst\b|\bcenters res excellence\b', affiliations_list[i])):
        ctrCount = ctrCount+1
        print('Research centers:', affiliations_list[i])
print('College/School:', collSchoolCount)
print('Research centers:', ctrCount)
```

```python
#Change authors with coll or School affiliation to Department affiliation
replaced_collSch = 0
for x in range(len(dataFrame)):
    dept_aff = []
    if re.search(r'coll|\bGeneral Studies College\b|School|\bsch engn\b', affiliations_list[x]):
        #print("issue", affiliations_list[x])
        for y in range(len(dataFrame)):
            if x!=y:
                if authors_list[x] == authors_list[y]:
                    #print('author found')
                    if affiliations_list[y] and affiliations_list[x]!=affiliations_list[y] and not re.search(';', affiliations_list[y]):
                        if re.search('Department', affiliations_list[y]):
                            if affiliations_list[y] not in dept_aff:
                                dept_aff.append(affiliations_list[y])

        for z in range(len(dept_aff)):
            if dept_aff[z] not in affiliations_list[x]:
                if z==0:
                    affiliations_list[x] = dept_aff[z]
                    replaced_collSch = replaced_collSch+1
                else:
                    affiliations_list[x] = affiliations_list[x] + " ; "+ dept_aff[z]
                        
print("College/School:", replaced_collSch)
#print(affiliations_list)
```
