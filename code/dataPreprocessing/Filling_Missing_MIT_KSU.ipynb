```python
import pandas as pd
import re
```

```python
#KSU collaboration dataset
#data = pd.read_csv(r'/Users/nadee/Documents/1Graduate Studies/Term 2/NLP/Project/Code/authors_all_publication_ksu_disambiguated.csv') 

#MIT collaboration dataset
data = pd.read_csv(r'/Users/nadee/Documents/1Graduate Studies/Term 2/NLP/Project/Code/authors_all_publication_mit_disambiguated.csv')
```

```python
df = pd.DataFrame(data, columns=['paper', 'title', 'author', 'affiliation', 'cite', 'year', 'area', 'reference_count', 'fund'])
```

```python
df.fillna('', inplace=True)
affiliations = df["affiliation"]
authors = df["author"]

affiliations_list = affiliations.tolist()
authors_list = authors.tolist()

for x in range(len(affiliations_list)):
    affiliations_list[x] = affiliations_list[x].title()
```

```python
len(df)
```

```python
#Filling missing affiliations with Department
filled_Dept = 0
for x in range(len(df)):
    missing_aff = []
    if not affiliations_list[x]:
        #print("issue", affiliations_list[x])
        for y in range(len(df)):
            if x!=y:
                if authors_list[x] == authors_list[y]:
                    if re.search('Department', affiliations_list[y]) and not re.search(';', affiliations_list[y]):
                        if affiliations_list[y] not in missing_aff:
                            missing_aff.append(affiliations_list[y])
        for z in range(len(missing_aff)):
            if missing_aff[z] not in affiliations_list[x]:
                if z==0:
                    affiliations_list[x] = missing_aff[z]
                    filled_Dept = filled_Dept+1
                else:
                    affiliations_list[x] = affiliations_list[x] + " ; "+ missing_aff[z]
                    
print("Filled with Department", filled_Dept)
```

```python
#Replace the multiple affiliations with the most frequent affiliation
import statistics 
from statistics import mode 
  
def most_common(List): 
    return(mode(List)) 

replaced_double_aff = 0

for x in range(len(df)):
    double_aff = []
    #Department, Division, Unit, Center
    if re.search(';', affiliations_list[x]) and re.search('Center', affiliations_list[x]): 
        for y in range(len(df)):
            if affiliations_list[y] and authors_list[x]==authors_list[y] and affiliations_list[x]!=affiliations_list[y] and re.search('Center', affiliations_list[y]):
                double_aff.append(affiliations_list[y])

        affiliations_list[x] = most_common(double_aff)
        replaced_double_aff = replaced_double_aff+1

#Department, Division, Unit, Center
print("Replaced double affiliation:", replaced_double_aff)
```
