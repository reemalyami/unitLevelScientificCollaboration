```python
import pandas as pd
import disamby.preprocessors as pre
from disamby import Disamby
```

```python
#data = pd.read_csv(r'/Users/nadee/Documents/1Graduate Studies/Term 2/NLP/Project/Code/KSU/all_authors_ksu_united.csv')
data = pd.read_csv(r'/Users/nadee/Documents/1Graduate Studies/Term 2/NLP/Project/Code/MIT/all_authors_mit_united_v0.csv')
```

```python
df = pd.DataFrame(data, columns=['paper', 'title', 'author', 'affiliation', 'cite', 'year', 'area', 'reference_count', 'fund'])
```

```python
authors = df['author']
affiliations = df['affiliation']

authors_list = authors.tolist()
affiliations_list = affiliations.tolist()

for x in range(len(authors_list)):
    authors_list[x] = authors_list[x].strip().lower()
    

#len(authors_list.unique())
```

```python
len(authors.unique())
```

```python
df = pd.DataFrame({
    'author': authors_list},
)
```

```python
pipeline = [
    pre.normalize_whitespace,
    pre.remove_punctuation,
    lambda x: pre.trigram(x) + pre.split_words(x)
]
```

```python
dis = Disamby(df, pipeline)
```

```python
unique_authors = dis.disambiguated_sets(threshold=0.5)
```

```python
for x in unique_authors:
    if len(x)>1:
        #print('Length:', len(x), "Unique:", x)
        #print(x, type(x))
        s = list(x) 
        #print("s", s)
        for i in range(len(s)):
            if i == 0:
                authors_list[s[i]] = authors_list[s[i]]
                print('Pre:', authors_list[s[i]], 'Post:', authors_list[s[i]])
            else:
                #print(s[i], authors_list[s[i]])
                print('Pre:', authors_list[s[i]])
                if affiliations_list[s[i]] == affiliations_list[s[i-1]]:
                    authors_list[s[i]] = authors_list[s[i-1]]
                    print('Post:', authors_list[s[i]])
```

```python
from pandas import DataFrame

author_df = DataFrame(authors_list, columns=['author'])
#print(len(author_df['author'].unique())) 7177
#print(df)

table = {"paper": [], "title": [], "author": [], "affiliation": [], "cite": [], "year": [], "area": [], "reference_count": [], "fund": []}

for x in range(len(df)):
    table["paper"].append(df['paper'][x])
    table["title"].append(df['title'][x])
    table["author"].append(author_df['author'][x])
    table["affiliation"].append(df['affiliation'][x])
    table["cite"].append(df['cite'][x])
    table["year"].append(df['year'][x])
    table["area"].append(df['area'][x])
    table["reference_count"].append(df['reference_count'][x])
    table["fund"].append(df['fund'][x])
```

```python
df=pd.DataFrame.from_dict(table, orient='index').transpose()
df.to_csv('authors_all_publication_mit_disambiguated.csv', index = False)
```
