```python
import pandas as pd
import re
```

```python
data = pd.read_csv("ksu_extracted_affiliation_manual.csv")
dictionary = pd.read_csv("ksu_dictionary_clean_manual.csv")
```

```python
data.head()
```

```python
data.head(10)
```

```python
data['o']= data['w']
```

```python
dictionary.head()
```

```python
#replace the abbreviations with the word from dictionary
for x in dictionary.index:
    abb = dictionary['abb'][x]
    word = dictionary['words'][x]
    regx = r"\b(?=\w){0}\b".format(abb)
    data['o'] = data['o'].replace({regx: word}, regex=True)
    
```

```python
data
```

```python
import string

def remove_punctuations(s):
    regex = re.compile('[%s]' % re.escape(string.punctuation))
    return regex.sub('', s)
```

```python
#cleaning from digits and punctuations

import re 

count = 0

for i in data.index:
    text = str(data['o'][i])
    #if the word contain digits remove it
    text = re.sub(r"\d", "", text)
    #print(text)
    #some unit values contains an address and hear I remove it
    unit = re.search(";",text)
    if unit:
        #print(unit)
        count += 1
        #print(unit.start())
        #print("before---",len(text),text)
        
        data['o'][i] = text[0: 0] + text[unit.start() + 1::]
        data['o'][i] = remove_punctuations(data['o'][i])
        #print("after---",len(text),text)
```

```python
data.to_csv("ksu_extracted_affiliation_dict_clean.csv", index=False)
```

```python
c = 0
d = 0
l = 0
div=0
u= 0
for i in df.index:
    words = str(df['o'][i]).strip().lower()
        
    if words.startswith('center'):
        c+=1
        words = words.replace('center', '')
        df['o'][i] = words.strip() + ' center'
        print(df['o'][i])
        
    if words.startswith('department'):
        d+=1
        words = words.replace('department', '')
        df['o'][i] = words.strip() + ' department'
        print(df['o'][i])
        
    if words.startswith('lab'):
        l+=1
        words = words.replace('lab', '')
        df['o'][i] = words.strip() + ' lab'
        print(df['o'][i])
    if words.startswith('division'):
        div+=1
        words = words.replace('division', '')
        df['o'][i] = words.strip() + ' division'
        print(df['o'][i])
    if words.startswith('unit'):
        u+=1
        words = words.replace('unit', '')
        df['o'][i] = words.strip() + ' unit'
        print(df['o'][i])
    else:
        df['o'][i] = str(df['o'][i]).strip()
```
