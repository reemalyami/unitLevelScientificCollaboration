```python
#check if the words are English or not 
import string

def isEnglish(s):
    return s.isascii()
```

```python
import re 

def remove_punctuations(s):
    regex = re.compile('[%s]' % re.escape(string.punctuation))
    return regex.sub('', s)

```

```python
def sequential_search(alist, item):
    position = 0

    while position < len(alist):
        if alist[position] == item:
            return position
        position = position + 1

    return None



#word=word.replace(w,"",1)
```

```python
dictionary = {"abb":[],"words":[]}
```

```python

count_abb = 0

for i in df.index:
    #split full names to list of words
    if df['w'][i] !=0:
        #list of words from extracted affiliations
        abbs = df['w'][i].split()
    if df['o'][i] !=0:   
        #list of words from retrived Google result 
        words = df['o'][i].split()
    else:
        continue
    for a in abbs:
        for w in words:
            a = str(a).strip().lower()
            w = str(w).strip().lower()
            
            #words with the same length are not abbreviations 
            if len(a)==len(w):
                continue
            #if the abbreviation than the word then skip
            elif len(a)>len(w):
                continue
            #if the first letter is not equal then skip
            elif a[0] != w[0]:
                continue #go to w+1
            else:
            #hear catch the words that start with similar letters
                check = check_equal_words(a,w)

                if check:
                    #print("YES!")
                    count_abb+=1
                    dictionary["abb"].append(remove_punctuations(a))
                    dictionary["words"].append(remove_punctuations(w))
    
```

```python
for i in range(count_abb):
    print(dictionary["abb"][i]," ",dictionary["words"][i])
```
